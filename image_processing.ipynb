```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

img = Image.open(r"blackpink.jpg").convert('L')  # Load in grayscale
img1 = Image.open(r"liverpool2324.jpg").convert('L')
img2 = Image.open(r"blackpink.jpg").convert('L')

img2 = img2.resize(img1.size)

img_np = np.array(img)
img1_np = np.array(img1)
img2_np = np.array(img2)

def show_image(img_np, title="Image"):
    plt.imshow(img_np, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# 1. Multiplication and Addition with a Constant
def multiply_add(img_np, alpha=1.5, beta=30):
    result = alpha * img_np + beta
    return np.clip(result, 0, 255).astype(np.uint8)

result_img_np = multiply_add(img_np)
show_image(result_img_np, "Multiplication and Addition")
```

```python
# 2. Multiplicative Gain (Scaling)
def apply_gain(img_np, gain=1.5):
    result = gain * img_np
    return np.clip(result, 0, 255).astype(np.uint8)

gained_img_np = apply_gain(img_np)
show_image(gained_img_np, "Gained Image")
```

```python
# 3. Linear Blend (Dyadic Operator)
def blend_images(img1_np, img2_np, alpha=0.5):
    blended = (1 - alpha) * img1_np + alpha * img2_np
    return np.clip(blended, 0, 255).astype(np.uint8)

blended_img_np = blend_images(img1_np, img2_np, alpha=0.5)
show_image(blended_img_np, "Blended Image")
```

```python
# 4. Invert Gamma Mapping
def invert_gamma(img_np, gamma=2.2):
    inv_gamma = 1.0 / gamma
    img_normalized = img_np / 255.0
    corrected_img = np.power(img_normalized, inv_gamma)
    return (corrected_img * 255).astype(np.uint8)

gamma_inverted_img_np = invert_gamma(img_np, gamma=2.2)
show_image(gamma_inverted_img_np, "Gamma Inverted Image")
```
